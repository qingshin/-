### 用户注册

- URL: /register/

- 方法: POST

- 描述: 用于新用户注册。

- 请求参数:

- username: 用户名

- email: 电子邮件

- password: 密码

- 返回:

- 成功: { "message": "User registered successfully" }

- 失败: { "error": "错误信息" }

### 用户登录

- URL: /login/

- 方法: POST

- 描述: 用户登录。

- 请求参数:

- username: 用户名

- password: 密码

- 返回:

- 成功: { "message": "Login successful", "token": "用户令牌" }

- 失败: { "error": "错误信息" }

### 用户登出

- URL: /logout/

- 方法: POST

- 描述: 用户登出。

- 请求参数: 无

- 返回:

- 成功: { "message": "Logout successful" }

### 获取用户信息

- URL: /user/<int:user_id>/

- 方法: GET

- 描述: 根据用户ID获取用户的基本信息。

- URL参数:

- user_id: 用户ID

- 返回:

- 成功: { "username": "用户名", "email": "电子邮件", "is_staff": "是否为管理员", "is_active": "是否激活" }

- 失败: { "error": "User not found" }

### 获取用户的关注列表

- URL: /user/<int:user_id>/following/

- 方法: GET

- 描述: 获取用户关注的其他用户列表。

- URL参数:

- user_id: 用户ID

- 返回:

- 成功: [{"username": "用户名", "email": "电子邮件"}, ...]

- 失败: { "error": "User not found" }

### 获取用户的粉丝列表

- URL: /user/<int:user_id>/followers/

- 方法: GET

- 描述: 获取关注用户的粉丝列表。

- URL参数:

- user_id: 用户ID

- 返回:

- 成功: [{"username": "用户名", "email": "电子邮件"}, ...]

- 失败: { "error": "User not found" }
