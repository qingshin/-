# 消息系统API需求文档

## 概述
本文档描述了消息系统的API接口，包括发送消息、接收消息、列出所有消息、搜索消息、获取消息详情、标记消息为已读/未读、删除消息和撤回消息等功能。

## API接口

### 发送消息

- **URL**: `/send_message/`
- **方法**: `POST`
- **参数**:
  - `sender`: 发送者用户ID
  - `receiver`: 接收者用户ID
  - `content`: 消息内容
- **返回**:
  - 成功: `{ "message": "Message sent successfully", "sent_message": { "id": 消息ID, "sender": 发送者ID, "receiver": 接收者ID, "content": 消息内容 } }`
  - 失败: `{ "error": "Method not allowed" }` (非POST请求)

### 接收消息

- **URL**: `/receive_messages/<user_id>/`
- **方法**: `GET`
- **参数**: 无（用户ID通过URL传递）
- **返回**: 消息列表，每个消息包含`id`, `sender_id`, `receiver_id`, `content`

### 列出所有消息

- **URL**: `/list_messages/`
- **方法**: `GET`
- **参数**: 无
- **返回**: 消息列表，每个消息包含`id`, `sender`, `receiver`, `content`

### 搜索消息

- **URL**: `/search_messages/<keyword>/`
- **方法**: `GET`
- **参数**: 无（搜索关键词通过URL传递）
- **返回**: 匹配关键词的消息列表，每个消息包含`id`, `sender`, `receiver`, `content`

### 获取消息详情

- **URL**: `/get_message_detail/<message_id>/`
- **方法**: `GET`
- **参数**: 无（消息ID通过URL传递）
- **返回**: 指定消息的详细信息，包括`id`, `sender`, `receiver`, `content`

### 标记消息为已读

- **URL**: `/mark_as_read/<message_id>/`
- **方法**: `POST`
- **参数**: 无（消息ID通过URL传递）
- **返回**:
  - 成功: `{ "message": "Message marked as read", "message_id": 消息ID }`
  - 失败: `{ "error": "Message not found" }`

### 标记消息为未读

- **URL**: `/mark_as_unread/<message_id>/`
- **方法**: `POST`
- **参数**: 无（消息ID通过URL传递）
- **返回**:
  - 成功: `{ "message": "Message marked as unread", "message_id": 消息ID }`
  - 失败: `{ "error": "Message not found" }`

### 删除消息

- **URL**: `/delete_message/<message_id>/`
- **方法**: `POST`
- **参数**: 无（消息ID通过URL传递）
- **返回**:
  - 成功: `{ "message": "Message deleted successfully", "message_id": 消息ID }`
  - 失败: `{ "error": "Message not found" }`

### 撤回消息

- **URL**: `/recall_message/<message_id>/`
- **方法**: `POST`
- **参数**: 无（消息ID通过URL传递）
- **返回**:
  - 成功: `{ "message": "Message recalled successfully", "message_id": 消息ID }`
  - 失败: `{ "error": "Message not found" }`

## 注意事项
- 所有POST请求需要CSRF豁免。
- 用户ID、消息ID和关键词需要通过URL路径传递。
- 返回的JSON数据中，成功响应将包含相关信息，失败响应将包含错误信息。
